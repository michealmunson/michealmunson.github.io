```python
import os
import pandas as pd
os.chdir('/Users/michealmunson/Documents/data-science/')

d = pd.read_csv('zhvi_state.csv')


def get_end_3_states(d):
    years = d.columns[5:]
    end_3_states = {}

    count = 0
    for i in years:
        end_3_states_temp = []
        end_3_states_temp += d.sort_values(years[count], ascending=True)['RegionName'][0:3].to_list()
        end_3_states_temp += d.sort_values(years[count], ascending=False)['RegionName'][0:3].to_list()
        end_3_states[i] = end_3_states_temp
        count += 1

    end_3_states = pd.DataFrame(end_3_states)
    cheapest_3_states = end_3_states.iloc[0:3]
    expensive_3_states = end_3_states.iloc[3:].reset_index(drop=True,inplace=False)
    ca_count = 0
    ma_count = 0
    dc_count = 0
    hi_count = 0
    nj_count = 0
    wv_count = 0
    ok_count = 0
    ky_count = 0
    ms_count = 0
    mi_count = 0
    for axis in range(0,3): 
        for name in expensive_3_states.iloc[axis]:
            if name == 'California':
                ca_count += 1
            elif name == 'Hawaii':
                hi_count += 1
            elif name == 'Massachusetts':
                ma_count += 1
            elif name == 'District of Columbia':
                dc_count += 1
            else:
                nj_count += 1
        for name in cheapest_3_states.iloc[axis]:
            if name == 'Oklahoma':
                ok_count += 1
            elif name == 'Kentucky':
                ky_count += 1
            elif name == 'Mississippi':
                ms_count += 1
            elif name == 'Michigan':
                mi_count += 1
            else:
                wv_count += 1


    expensive_state_freq = {
        "CA":[ca_count],
        "MA":[ma_count],
        'DC':[dc_count],
        'HI':[hi_count],
        'NJ':[nj_count]
    }
    cheapest_state_freq = {
        "WV":[wv_count],
        'OK':[ok_count],
        'KY':[ky_count],
        'MS':[ms_count],
        'MI':[mi_count]
    }

    return pd.DataFrame(expensive_state_freq), pd.DataFrame(cheapest_state_freq)

expensive_states, poor_states = get_end_3_states(d)
print(pd.DataFrame(expensive_states))



```
